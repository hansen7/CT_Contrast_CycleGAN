```python
im1 = Image.open(os.path.join(result_dir, 'epoch200_fake_A.png'))#.convert('L')
im2 = Image.open(os.path.join(result_dir, 'epoch200_idt_A.png'))#.convert('L')
im3 = Image.open(os.path.join(result_dir, 'epoch200_real_B.png'))#.convert('L')
im4 = Image.open(os.path.join(result_dir, 'epoch200_fake_B.png'))#.convert('L')
# im4 = Image.open(os.path.join(result_dir, 'epoch200_real_B.png'))#.convert('L')

plt.figure(figsize=(16,3))
plt.subplot(141); plt.imshow(im1, cmap='gray')
plt.subplot(142); plt.imshow(im2, cmap='gray')
plt.subplot(143); plt.imshow(im3, cmap='gray')
plt.subplot(144); plt.imshow(im4, cmap='gray')
```

```python
epochs = list(set([img.split('_')[0] for img in os.listdir(result_dir)]))
```

```python
len(epochs)
```

```python
for epoch in np.sort(epochs)[60:70]:
    # print(epoch)
    
    im1 = Image.open(os.path.join(result_dir, '%s_real_A.png' % epoch))  # true contrast images
    im2 = Image.open(os.path.join(result_dir, '%s_fake_B.png' % epoch))  # generated non-contrast images
    im3 = Image.open(os.path.join(result_dir, '%s_fake_A.png' % epoch))  # generated contrast images
    im4 = Image.open(os.path.join(result_dir, '%s_real_B.png' % epoch))  # true non-contrast images

    plt.figure(figsize=(16,3))
    plt.subplot(141); plt.imshow(im1, cmap='gray')
    plt.title("true contrast images")
    
    plt.subplot(142); plt.imshow(im2, cmap='gray')
    plt.title("generated non-contrast images")
    
    plt.subplot(143); plt.imshow(im3, cmap='gray')
    plt.title("generated contrast images")
    
    plt.subplot(144); plt.imshow(im4, cmap='gray')
    plt.title("true non-contrast images")
    
    plt.suptitle(epoch)
    plt.show()
    plt.close()
```
